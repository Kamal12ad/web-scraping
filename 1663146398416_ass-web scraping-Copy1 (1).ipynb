```python
import requests 
from bs4 import BeautifulSoup
```

# scraping websites

```python
url="https://www.imdb.com/list/ls073043203/"
```

```python
url
```

```python
res=requests.get(url)
```

```python
res
```

# parsing

```python
soup=BeautifulSoup(res.text)
```

```python
soup
```

# to select aparticular portion

```python
cartoons=soup.select('.lister-item-header a')
```

```python
cartoons
```

```python
cartoons_title=[]
cartoons_links=[]
```

```python
for item in cartoons:

      cartoons_title.append(item.text)
       
```

```python
links="https://imdb.com"+item["href"]

cartoons_links.append(links)
```

```python
links
```

```python
links
```

```python
year=soup.select('.text-muted.unbold')
```

```python
year
```
